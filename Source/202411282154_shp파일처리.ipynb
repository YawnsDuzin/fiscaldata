```python
import geopandas as gpd
import folium
```

```python
gdf = gpd.read_file(path_dateset + file_name, encoding="cp949")
```

```python
gdf = gdf.to_crs(epsg=4326)  # WGS84 좌표계로 변환
```

```python
m = folium.Map(location=[gdf.centroid.y.mean(), gdf.centroid.x.mean()], zoom_start=6)
```

```python
folium.GeoJson(gdf).add_to(m)
```
