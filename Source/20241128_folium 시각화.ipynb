```python
file_name = '원본/법정구역 GeoJSON 데이터_23년8월/법정구역_시군구_simplified.geojson'
```

```python
# 자치구 정보 (geojson)
def seoul_municipalities_geo_simple():
    geo_path = path_dateset + file_name
    # encoding을 UTF-8로 지정
    with open(geo_path, 'rt', encoding='utf-8') as f:
        geo_str2 = json.load(f)
    return geo_str2
```

```python
_geo_data = seoul_municipalities_geo_simple()
```

```python
# folium지도 시각화 관련 - 자치구별 밀도 표시
def folium_gu_density(_geo_data, _data, _name, _fill_color = 'Blues', fill_opacity = 0.7):
    choropleth = folium.Choropleth(geo_data = _geo_data,
                  #data = _data,
                  name = _name,
                  key_on = 'feature.properties.SIG_KOR_NM',
                  fill_color = _fill_color,
                  fill_opacity=fill_opacity              # 채우기 불투명도
                )
    return choropleth
```

```python
import folium
import geopandas as gpd

# GeoJSON 파일 로드
gdf = seoul_municipalities_geo_simple()

# 중심 좌표 설정 (대한민국 중심 좌표)
center = [36.5, 127.5]

# 지도 생성
m = folium.Map(location=center, zoom_start=7)

# 시군구별로 색상 지정
colors = ['red', 'blue', 'green', 'purple', 'orange', 'darkred', 'lightred', 'beige', 'darkblue', 'darkgreen']

# 시군구 경계 그리기
for idx, row in enumerate(gdf):
    sim = folium.GeoJson(
        row['geometry'],
        style_function=lambda x: {
            'fillColor': colors[idx % len(colors)],
            'color': 'black',
            'weight': 1,
            'fillOpacity': 0.7
        }
    )
    folium.Popup(row['SIG_KOR_NM']).add_to(sim)
    sim.add_to(m)

# 레이어 컨트롤 추가
folium.LayerControl().add_to(m)

# 지도 저장
m.save('korea_sigungu_map.html')
import webbrowser
webbrowser.open("korea_sigungu_map.html")
```

```python
import numpy as np

save_name = 'save_file'

center = [37.5665, 126.9780]
bike_map = folium.Map(location=center, zoom_start=12, tiles='openstreetmap', name='OpenStreetMap')

# folium_gu_density : 자치구별 밀집도 표시
choropleth1 = folium_gu_density(seoul_municipalities_geo_simple(), np.NaN, '자전거도로_총계 연장(㎞)_밀집도', 'Blues', 0.7)
choropleth1.add_to(bike_map)

#folium.TileLayer('openstreetmap', name='OpenStreetMap').add_to(bike_map)
folium.TileLayer('cartodbpositron', name='CartoDB Positron').add_to(bike_map)
folium.TileLayer('cartodbdark_matter', name='CartoDB Dark Matter').add_to(bike_map)

folium.LayerControl().add_to(bike_map)

# web에서 표시하기
# 지도 저장
bike_map.save(save_name + ".html")
import webbrowser
webbrowser.open(save_name + ".html")
```
