### 저장 데이터 불러오기

```python
df_temp = pd.read_csv(path_dateset + '전처리/재정자주도_자립도_통합재정수지비율(병합).csv')
df_temp.head(3)
```

```python
df_temp.shape
```

```python
import plotly.graph_objects as go
import plotly.express as px

sort_val = '통합재정수지비율[당초]'
df_1 = df_temp.sort_values(by = sort_val, ascending = False).head(20)   # 상위20
df_2 = df_temp.sort_values(by = sort_val, ascending = True).head(20)    # 하위20

df_1[sort_val] = df_1[sort_val].apply(lambda x: round(x, 2))
df_2[sort_val] = df_2[sort_val].apply(lambda x: round(x, 2))
# ===========================================================

# 그라데이션 색상 생성
values = df_1[sort_val]
colorscale = px.colors.sequential.Viridis
norm_values = (values - values.min()) / (values.max() - values.min())
colors = [colorscale[int(v * (len(colorscale) - 1))] for v in norm_values]

# 라인 플롯
fig1 = go.Figure()
fig1.add_trace(
    go.Scatter(
        x=df_1['자치단체'],
        y=df_1[sort_val],
        mode='lines+markers',
        line=dict(
            color='lightgrey',
            width=2
        ),
        marker=dict(
            color=colors,
            size=10,
            line=dict(
                color='white',
                width=1
            )
        )
    )
)

# 라인 플롯에 화살표와 값 추가
for i in range(len(df_1)):
    fig1.add_annotation(
        x=df_1['자치단체'].iloc[i],
        y=df_1[sort_val].iloc[i],
        text=f'{df_1[sort_val].iloc[i]}%',
        showarrow=False,
        yshift=10
    )

fig1.update_layout(
    title='자치단체별 ' + sort_val + ' (상위20) 비교',
    xaxis_title='자치단체',
    yaxis_title=sort_val,
    font=dict(family='Malgun Gothic'),
    showlegend=False
)

# ===========================================================
# 바 플롯
fig2 = go.Figure()
fig2.add_trace(
    go.Bar(
        x=df_2['자치단체'],
        y=df_2[sort_val],
        marker=dict(
            color=df_2[sort_val],
            colorscale='Viridis'
        ),
        text=df_2[sort_val].apply(lambda x: f'{x:.2f}%'),
        textposition='outside',
        textfont=dict(size=12)
    )
)

fig2.update_layout(
    title='자치단체별 ' + sort_val + ' (하위20) 비교',
    xaxis_title='자치단체',
    yaxis_title='sort_val',
    font=dict(family='Malgun Gothic'),
    showlegend=False
)

# ===========================================================
# 그래프 표시
fig1.show()
fig2.show()
```
